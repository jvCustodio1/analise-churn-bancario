```python
import pandas as pd

# Adicionamos o parâmetro encoding='latin1'
df = pd.read_csv('Dados_clientes.csv', encoding='latin1')

# Ver as primeiras linhas
display(df.head())
```

```python
print(f"O banco de dados tem {df.shape[0]} linhas e {df.shape[1]} colunas.")
```

```python
print(df.isnull().sum())
```

```python
ativos = df[df['Categoria'] == 'Cliente'] #CRIAR VARIAVEL PARA CLIENTES ATIVOS - SELECIONOU A CATEGORIA CLIENTES E DEIXOU COMO CLIENTE
cancelados = df[df['Categoria'] == 'Cancelado'] #CRITOU VARIAVEL PARA CLIENTES CANCELADOS - SELECIONOU A COLUNA CATEGORIA E = A CANCELADO
print(f"Clientes Ativos: {ativos}")
print(f"Clientes Cancelados: {cancelados}")
```

```python
contagem_clientes = df['Categoria'].value_counts()

print(contagem_clientes)
```

```python
# O cálculo precisa estar dentro das chaves {} para o Python entender que deve executar a conta
print(f"A porcentagem de clientes cancelados é de: {(cancelados / ativos) * 100}%")
```

```python
# 1. Pegando apenas a QUANTIDADE (o número) de cada um
ativos = df[df['Categoria'] == 'Cliente'].shape[0]
cancelados = df[df['Categoria'] == 'Cancelado'].shape[0]

# 2. Calculando o total
total = ativos + cancelados

# 3. Agora sim, o print com o cálculo matemático
print(f"Total de clientes: {total}")
print(f"Ativos: {ativos} | Cancelados: {cancelados}")
print(f"A porcentagem de cancelamento (Churn) é de: {(cancelados / total) * 100:.2f}%")
```

```python
# 1. Separamos a tabela original em duas: uma só com ativos e outra só com cancelados
# Isso facilita comparar o comportamento dos dois grupos isoladamente.
df_ativos = df[df['Categoria'] == 'Cliente'] # Criou uma variavel chamada df_ativos que é somente para uma tabela de clientes ativos 
df_cancelados = df[df['Categoria'] == 'Cancelado'] # Criou uma variavel chamada df_cancelados que é somente para tabela de clientes cancelados

# 2. Vamos calcular a média de 'Contatos nos últimos 12 meses' para cada grupo
# O .mean() soma todos os valores da coluna e divide pela quantidade de linhas.
media_contatos_ativos = df_ativos['Contatos 12m'].mean()
media_contatos_cancelados = df_cancelados['Contatos 12m'].mean()

# 3. Vamos calcular a média de 'Inatividade nos últimos 12 meses'
# Queremos saber se quem cancela costuma ficar mais tempo "parado" antes de sair.
media_inatividade_ativos = df_ativos['Inatividade 12m'].mean()
media_inatividade_cancelados = df_cancelados['Inatividade 12m'].mean()

# Exibindo os resultados para comparação
print(f"Média de Contatos (Ativos): {media_contatos_ativos:.2f}")
print(f"Média de Contatos (Cancelados): {media_contatos_cancelados:.2f}")
print("-" * 30)
print(f"Média de Inatividade (Ativos): {media_inatividade_ativos:.2f}")
print(f"Média de Inatividade (Cancelados): {media_inatividade_cancelados:.2f}")
```

```python
# 1. Criamos uma tabela de frequência para a Faixa Salarial dos dois grupos
# O normalize=True transforma os números brutos em porcentagem (0.15 vira 15%)
perfil_renda_ativos = df_ativos['Faixa Salarial Anual'].value_counts(normalize=True) * 100
perfil_renda_cancelados = df_cancelados['Faixa Salarial Anual'].value_counts(normalize=True) * 100

# 2. Exibindo os resultados
print("Distribuição de Renda - Clientes ATIVOS:")
print(perfil_renda_ativos.map("{:.2f}%".format)) # Formata cada linha com 2 casas e %

print("\n" + "="*40 + "\n")

print("Distribuição de Renda - Clientes CANCELADOS:")
print(perfil_renda_cancelados.map("{:.2f}%".format))
```

```python
# 1. Criamos uma tabela cruzada (crosstab)
# Ela mostra exatamente quantos clientes de cada cartão cancelaram ou ficaram.
# O 'margins=True' adiciona uma linha de "Total" no final para facilitar.
comparativo_cartao = pd.crosstab(df['Categoria'], df['Categoria Cartão'], margins=True) #pd.crosstab: É o "Tabela Dinâmica" do Excel, mas no Python. Ele cruza duas informações e conta quantas vezes elas aparecem juntas.

# 2. Vamos calcular a porcentagem de cancelamento dentro de cada tipo de cartão
# Para isso, pegamos a linha 'Cancelado' e dividimos pela linha 'All' (Total)
# O .loc permite que a gente selecione linhas específicas pelo nome.
taxa_por_cartao = (comparativo_cartao.loc['Cancelado'] / comparativo_cartao.loc['All']) * 100 #.loc['Nome']: É o jeito de você dizer ao Python: "Olhe especificamente para a linha que tem esse nome".

print("Quantidade bruta por categoria de cartão:")
print(comparativo_cartao)

print("\n" + "-"*40 + "\n")

print("Taxa de Cancelamento por tipo de cartão:")
print(taxa_por_cartao.map("{:.2f}%".format))
```

```python
import plotly.express as px # Importando a biblioteca de gráficos

# 1. Vamos criar um gráfico de barras comparando as categorias
# Usamos o 'df' original para que o gráfico conte automaticamente
fig = px.histogram(df, 
                   x="Categoria Cartão", 
                   color="Categoria", 
                   barmode="group",
                   title="Comparativo de Ativos vs Cancelados por Tipo de Cartão")

# 2. Exibir o gráfico
# No VS Code, isso abrirá uma aba no seu navegador ou uma janela ao lado
fig.show()
```

```python
# 1. Criamos um histograma para comparar a quantidade de produtos contratados
# Queremos ver se quem tem 1 ou 2 produtos cancela mais do que quem tem 5 ou 6.
fig_produtos = px.histogram(df, 
                            x="Produtos Contratados", 
                            color="Categoria", 
                            barmode="group",
                            title="Cancelamento por Quantidade de Produtos Contratados")

# 2. Exibindo o novo gráfico
fig_produtos.show()

# 3. Calculando a média exata para o seu relatório
media_prod_ativos = df_ativos['Produtos Contratados'].mean()
media_prod_cancelados = df_cancelados['Produtos Contratados'].mean()

print(f"Média de produtos dos Ativos: {media_prod_ativos:.2f}")
print(f"Média de produtos dos Cancelados: {media_prod_cancelados:.2f}")
```

```python
# 1. Calculando as médias com os nomes corretos das colunas
# Note que usei 'Limite' e 'Limite Consumido' conforme sua imagem
media_limite_ativos = df_ativos['Limite'].mean()
media_limite_cancelados = df_cancelados['Limite'].mean()

# 2. Gráfico de Dispersão corrigido
# Troquei 'Consumo Medio' por 'Limite Consumido'
fig_limite = px.scatter(df, 
                        x="Limite", 
                        y="Limite Consumido", 
                        color="Categoria",
                        title="Relação: Limite Total vs Limite Consumido",
                        hover_data=['Idade', 'Faixa Salarial Anual'])

fig_limite.show()

print(f"Média de Limite Total (Ativos): R${media_limite_ativos:.2f}")
print(f"Média de Limite Total (Cancelados): R${media_limite_cancelados:.2f}")
```

```python
# Comparando a taxa de utilização média
utilizacao_ativos = df_ativos['Taxa de Utilização Cartão'].mean()
utilizacao_cancelados = df_cancelados['Taxa de Utilização Cartão'].mean()

print(f"Taxa de Utilização Média (Ativos): {utilizacao_ativos:.2f}")
print(f"Taxa de Utilização Média (Cancelados): {utilizacao_cancelados:.2f}")
```
